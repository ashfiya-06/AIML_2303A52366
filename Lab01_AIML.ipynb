# **Part 1 – Implement Basic Data Structures using Numpy, Pandas**

Lists, *Arrays*

```python
L1 = [0, -1, -5, 10]
L2 = [-10, 5, 1, 29]
A1 = np.array([1, 3, 5, 7])
A2 = np.array([2, 4, 6, 8])
print(L1)
print(A1)
print(A2)
print(L2)
```

Their types~

```python
print("L1 type is: ", type(L1))
print("A1 type is: ", type(A1))
print("L2 type is: ", type(L2))
print("A2 type is: ", type(A2))
```

Mathematical Operations

```python
print(A1*A2)
print(A1*A1)
print(A2*A2)
print(A1/A1)
print(A2/A2)
print(A1/A2)
print(A1**A2)
```

Combining text with Numpy fuction to generate a textual output

```python
print(f"Adding two numpy arrays {A1} and {A2} together: ", A1+A2)
```

Implementing sin, log, exp:

```python
print("Sine of A1: ",np.sin(A1))
print("Sine of A2: ",np.sin(A2))
print("Log of A1: ",np.log(A1))
print("Log of A2: ",np.log(A2))
print("Exp of A1: ",np.exp(A1))
print("Exp of A2: ",np.exp(A2))
print("Log2 of A1: ",np.log2(A1))
print("Log2 of A2: ",np.log2(A2))
```

# **Part 2 – Visualization of Data using matplotlib, pyplots Packages[CO2]**


 1. Generating a Data set for Health Care

```python
people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
```

```python
age =[25, 30, 35, 40]
plt.scatter(age, height)
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Age vs Height")
plt.show()
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python
plt.bar(people, age)
plt.xlabel("People")
plt.ylabel("Age")
plt.title("Age of People")
plt.show()
```

```python
plt.bar(people, height, color='cyan', edgecolor = 'black')
plt.xlabel("People")
plt.xticks(fontsize=14,rotation=30)
plt.ylabel("Height")
plt.yticks(fontsize=14)
plt.title("Height of People")
plt.show()
```

```python
plt.hist(weight, color='palegreen', edgecolor='black')
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.title("Distribution of Weights")
plt.gca().set_facecolor('lightblue')
plt.show()
```
